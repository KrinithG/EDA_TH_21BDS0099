# **Merging dataframes using Join**

```python
# Splitting the dataset into 4 dataframes
dfkg1 = krinith_df.iloc[: len(krinith_df)//4]
dfkg2 = krinith_df.iloc[len(krinith_df)//4:len(krinith_df)//2]
dfkg3 = krinith_df.iloc[len(krinith_df)//2:3*len(krinith_df)//4]
dfkg4 = krinith_df.iloc[3*len(krinith_df)//4:]
```

```python
#Using Concat()
dfkg22 = pd.concat([dfkg1, dfkg2], ignore_index=True)
dfkg44 = pd.concat([dfkg3, dfkg4], ignore_index=True)
```

# **Outlier detection and filtering**

```python
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = krinith_df['nr'].quantile(0.25)
Q3 = krinith_df['nr'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

#Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = krinith_df[(krinith_df['nr'] < lower_bound) | (krinith_df['nr'] > upper_bound)]

print(outliers[['nr']])

df = krinith_df[(krinith_df['nr'] >= lower_bound) & (krinith_df['nr'] <= upper_bound)]

print(df[['nr']])
```

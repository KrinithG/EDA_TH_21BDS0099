# **Outlier detection and filtering**

```python
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = krinith_df['nr'].quantile(0.25)
Q3 = krinith_df['nr'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

#Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = krinith_df[(krinith_df['nr'] < lower_bound) | (krinith_df['nr'] > upper_bound)]

print(outliers[['nr']])

df = krinith_df[(krinith_df['nr'] >= lower_bound) & (krinith_df['nr'] <= upper_bound)]

print(df[['nr']])
```

## **1-D Statistical Analysis**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
mean_krinith= np.mean(krinith_df.wage)
median_krinith = np.median(krinith_df.wage)
```

```python
variance_krinith = np.var(krinith_df.wage, ddof=1)
std_krinith= np.std(krinith_df.wage, ddof=1)
skewness_krinith = stats.skew(krinith_df.wage)
kurtosis_krinith = stats.kurtosis(krinith_df.wage)
```

```python
print(f"Mean: {mean_krinith:.2f}")
print(f"Median: {median_krinith}")
print(f"Variance: {variance_krinith:.2f}")
print(f"Standard Deviation: {std_krinith:.2f}")
print(f"Skewness: {skewness_krinith:.2f}")
print(f"Kurtosis: {kurtosis_krinith:.2f}")
```

```python
# Histogram Plot
plt.hist(krinith_df.wage, bins=7, edgecolor='black', alpha=0.7)
plt.axvline(mean_krinith, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean_krinith:.2f}")
plt.axvline(median_krinith, color='blue', linestyle='dashed', linewidth=2, label=f"Median: {median_krinith}")
plt.xlabel("Wages")
plt.ylabel("Frequency")
plt.title("1-D Statistical Analysis Histogram")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Covariance & Correlation
df = krinith_df[['nr', 'wage']]
cov_matrix = df.cov()
corr_matrix = df.corr()
print("\n Covariance Matrix:\n", cov_matrix)
print("\n Correlation Matrix:\n", corr_matrix)
```

```python
# Linear Regression (Best-Fit Line)
X = krinith_df['nr'].values.reshape(-1, 1)
Y = krinith_df['wage'].values.reshape(-1, 1)
regressor = LinearRegression()
regressor.fit(X, Y)
predicted_Y = regressor.predict(X)
slope, intercept = regressor.coef_[0][0], regressor.intercept_[0]
print(f"\n Regression Line: {'nr'} = {slope:.2f}  {'wage,'} + {intercept:.2f}")
```
